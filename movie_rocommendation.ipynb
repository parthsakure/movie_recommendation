```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer()

data = pd.read_csv('movie_dataset.csv')
data.dtypes
```

```python
features = ['keywords','cast','genres','director']
# features = ['genres']
data.isnull().sum()
```

```python
for i in features:
    data[i] = data[i].fillna('')
data.isnull().sum()
```

```python
def combine(a):
    s= ''
    for i in features:
        s += str(a[i]) + " "
    s = s[:-1]
    return s
data['features'] = data.apply(combine, axis=1)
data.head(3)
```

```python
cv.fit(data['features'])
voc = cv.vocabulary_
matrix = cv.transform(data['features'])
cos_sim = cosine_similarity(matrix)
# voc
```

```python
movie_name = input("Enter Movie name:")

movie_num = data[data.title==movie_name]['index'].values[0]

movies = sorted(list(enumerate(cos_sim[movie_num])), key=lambda x: x[1], reverse=True)[1:]
recommended_movies = []
for i in range(5):
    recommended_movies.append(data[data.index == movies[i][0]]['original_title'].values[0])

print("Top 5 Recommended Movies for `{}` are: ".format(movie_name))
for i,val in enumerate(recommended_movies):
    print(i+1,'. ',val)
```
